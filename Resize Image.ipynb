```python
import os
import glob
from tqdm import tqdm
from PIL import Image, ImageFile
from joblib import Parallel, delayed
ImageFile.LOAD_TRUNCATED_IMAGES = True
```

```python
def resize_image(image_path, output_folder, resize):
    base_name = os.path.basename(image_path)
    outpath = os.path.join(output_folder, base_name)
    img = Image.open(image_path)
    img = img.resize((resize[1], resize[0]), resample=Image.BILINEAR)
    img.save(outpath)
```

```python
#Resize Train Images
input_folder = "S:/Skin Cancer Detection/data/jpeg/train/"
output_folder = "S:/Skin Cancer Detection/data/jpeg/train_resize/"
images = glob.glob(os.path.join(input_folder, "*.jpg")) 

# glob to make list of jpeg files
# i is from for loop , looping over image paths
Parallel(n_jobs = 5)(delayed(resize_image)(i, output_folder, (512,512)) for i in tqdm(images))
```

```python
#Resize Test Images
input_folder = "S:/Skin Cancer Detection/data/jpeg/test/"
output_folder = "S:/Skin Cancer Detection/data/jpeg/test_resize/"
images = glob.glob(os.path.join(input_folder, "*.jpg")) # glob to make list of jpeg files
Parallel(n_jobs = 12)(
    delayed(resize_image)(
        i,
        output_folder,
        (512,512)
    ) for i in tqdm(images)
)
```
